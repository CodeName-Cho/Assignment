## fig,ax 설정과 그냥 냅다 plt 함수의 차이점
fig, ax를 설정하는 방식과 plt 함수를 바로 사용하는 방식은 각각 장단점과 쓰임새가 다릅니다. 이를 구체적으로 설명하면:

**1. fig, ax = plt.subplots() 방식**

이 방식은 객체 지향적(Object-Oriented, OO) 접근입니다. fig와 ax는 각각:

fig: 그림 전체(Figure)를 나타내며, 하나의 그림 안에 여러 개의 그래프(subplots)를 포함할 수 있습니다.

ax: 실제 그래프가 그려질 축(Axes)을 의미합니다. 축은 독립적으로 스타일을 지정하거나 여러 개의 그래프를 그릴 때 사용됩니다.

. - - - - - - - - - - - - - - - - - - - - - - - - - -

장점:
복잡한 그래프를 그릴 때 유리: 여러 개의 서브플롯(subplots)을 사용하거나 그래프 요소(축, 범례, 타이틀)를 세밀하게 제어할 때 유용합니다.

세밀한 제어 가능: 각 축에 대한 세부 설정이나 커스터마이징이 필요할 때, 예를 들어 각 축의 레이블, 제목, 스케일을 따로 설정하고 싶을 때 적합합니다.

여러 서브플롯: 한 번에 여러 개의 그래프를 그려야 할 때, 객체 지향 접근을 쓰면 더 간단하고 직관적입니다.

---

**2. plt.plot()과 같은 함수형 접근**

이 방식은 절차적(Procedural) 접근입니다. plt 함수는 단순하게 한 번에 그래프를 그리는 방식으로, 코드가 짧고 간단한 작업에 적합합니다.

. - - - - - - - - - - - - - - - -

장점:

빠르고 간단: 기본적인 그래프를 그릴 때 적합합니다. 그래프의 세부 요소를 다루는 일이 많지 않다면 plt 함수만으로도 충분합니다.

빠른 시각화: 간단한 시각화가 필요할 때 빠르게 사용할 수 있습니다.

@ 6. 그래프 그리기

```python
import matplotlib.pyplot as plt
fig, ax  = plt.subplots()
ax = closing_year.plot()
plt.show

fig, ax = plt.subplots()
ax = closing_year_q.plot()
plt.show()
```

- first_valid_index : 가장 최근의 데이터를 가져옴

### apply

apply() 함수는 데이터프레임의 행(axis=1) 또는 열(axis=0)에 함수를 적용할 수 있습니다.

열이나 행에 함수를 적용할 때 유연하게 다양한 함수를 사용할 수 있고, 원소 단위로 함수가 작동합니다.

### pd.Series
pd.Series를 이용하는 이유는 주로 pandas의 데이터 구조에서 각 열이나 행을 개별적으로 처리할 수 있기 때문입니다

각 열의 데이터 타입에 맞는 기능 제공: Series는 단일 데이터 타입(숫자, 문자열 등)을 갖는 1차원 배열 구조로, 열마다 특화된 처리를 할 수 있게 함. DataFrame의 경우 다양한 데이터 타입을 포함 가능하지만, Series는 각 열에 더 적합한 처리를 제공할 수 있음.

.

@ 7. last_valid_index, first_valid_index 메서드를 사용해 각 나라의 에볼라 발병일 구하기

```python
last_valid = ebola.apply(pd.Series.last_valid_index)
print(last_valid)
```

```python
first_valid = ebola.apply(pd.Series.first_valid_index)
print(first_valid)
```

@ 8. 발병일 동일한 출발선으로 옮기려면 처음 발병일에서 각 나라 발병일 뺀 만큼 (=shif_values 만큼) 옮기면 됨

```python
earliest_date = ebola.index.min()
print(earliest_date)
```

```python
shift_values = last_valid - earliest_date
print(shift_values)
```

@ 9. shift 메서드 이용해 모든 열 값을 shift_values 값만큼 옮기기

```python
ebola
```

```python
ebola_dict = {}
for idx, col in enumerate(ebola):
  d = shift_values[idx].days
  shifted = ebola[col].shift(d)
  ebola_dict[col] = shifted
```

@ 10. 시간 다시 설정한 애가 딕셔너리 형태라 데이터프레임으로 변환

```python
ebola_shift = pd.DataFrame(ebola_dict)
```

@ 11. 이제 최초 발병일 기준으로 옮겨짐

```python
print(ebola_shift.tail())
```

### .drop()

() 내의 애들 삭제

@ 12. 인덱스를 Day 열로 지정하고 그래프에 필요 없는 Date, Day 열은 삭제하면 그래프 위한 데이터프레임 완성

```python
ebola_shift.index = ebola_shift['Day']
ebola_shift = ebola_shift.drop(['Date', 'Day'], axis=1) # axis=1: 열을 삭제하겠다는 의미

print(ebola_shift.tail())
```
